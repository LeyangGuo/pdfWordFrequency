### 以下是用于对文本进行处理,统计的函数
请确保已经执行过,通常不需要修改

```python
import pandas as pd
import re
```

```python
#预处理文本,便于统计
def textWash(text):
    #处理连词符
    connPatt = re.compile(r'-\n')
    text = re.sub(connPatt,"",text)
        
    #全部转换为小写
    text = text.lower()
    
    #将所有非英文小写字母剔除
    noneEnPatt = re.compile(r'[^a-z]')
    text = re.sub(noneEnPatt," ",text)
    
    #剔除所有单个英文字符
    singleCharPatt = re.compile(r'\b[a-z]\b')
    text = re.sub(singleCharPatt," ",text)
    
    return text
```

```python
#将文本划分为单词列表
def wordsSplit(text):
    #简单地根据空格划分
    return text.split()
```

```python
#统计单词
def wordsCount(wordsList):
    #空词典,用于保存结果
    counts = {}
    
    #对列表中的单词计数
    for word in wordsList:
        counts[word] = counts.get(word,0) + 1
    
    return counts
```

```python
#把词按频率排序
def countsSort(counts):
    #按字典的value排序
    return sorted(counts.items(), key = lambda kv:(kv[1], kv[0]), reverse = True)
```

### 在下面的单元导入txt

```python
fileName = "tlp.txt" # txt文件名
txtFile = open(fileName,'r')
text = txtFile.read()
```

### 完成统计,结果保存在res中

```python
res = wordsCount( wordsSplit( textWash(text)))
```

```python
countsSort(res)
```

### 将结果导出到cvs文件中

```python
#将结果转换为pandas的DataFrame结构
df = pd.DataFrame.from_dict(res, orient='index',columns=['Frequency'])

#cvs文件名与txt相同
csvName = fileName.replace(".txt","") + ".csv"
#写入cvs
df.to_csv(csvName)
```
